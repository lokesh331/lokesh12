```python
file_path='/content/Movie.csv'
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
movies_df=pd.read_csv(file_path)
```

```python
movies_df[0:5]
```

```python
len(movies_df.movie.unique())
```

```python
user_movies_df=movies_df.pivot(index='userId',columns='movie',values='rating')
```

```python
user_movies_df
```

```python
user_movies_df.index=movies_df.userId.unique()
```

```python
user_movies_df.fillna(0,inplace=True)
```

```python
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine,correlation
from scipy.spatial.distance import  correlation
ok_1=[2,4,0,8]
ok_2=[2,0,0,4]
correlation_distance = correlation(ok_1,ok_2)
print(correlation_distance)
```

```python
user_sim = 1-pairwise_distances(user_movies_df.values,metric='cosine')
user_sim
```

```python
user_sim_df=pd.DataFrame(user_sim)
```

```python
print(user_sim_df)
```

```python
user_sim_df.index=movies_df.userId.unique()
user_sim_df.columns=movies_df.userId.unique()
```
